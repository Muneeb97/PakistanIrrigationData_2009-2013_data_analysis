<b>Importing Basic Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="dark")
sns.set(font_scale=1.5)
%matplotlib inline
```

<b>Reading Data From a CSV File Format<b>

```python
df = pd.read_csv('Data.csv')
```

<b>Checking The Basic View Of The DataFrame

```python
df.head()
```

<b>Checking The Basic Info Of The DataFrame

```python
df.info()
```

<b>Getting The Total Sum Of Values Per Year

```python
df_sum_year = df.groupby('Year').sum()
df_sum_year
```

<b>Total Area Irrigated in Pakistan Per Year Visualized

```python
plt.barh(df_sum_year.index,df_sum_year['Total'])
plt.title('Total Area Irrigated in Pakistan')
plt.xlabel('Area in million hectares)')
plt.ylabel('Year')
```

<b>Total Area Irrigated In Each Province Per Year

```python
kpk_year = df[df['Province'] == 'Khyber Pakhtunkhwa']['Year']
kpk_total = df[df['Province'] == 'Khyber Pakhtunkhwa']['Total']

punjab_year = df[df['Province'] == 'Punjab']['Year']
punjab_total = df[df['Province'] == 'Punjab']['Total']

balochistan_year = df[df['Province'] == 'Balochistan']['Year']
balochistan_total = df[df['Province'] == 'Balochistan']['Total']

sindh_year = df[df['Province'] == 'Sindh']['Year']
sindh_total = df[df['Province'] == 'Sindh']['Total']
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2,figsize=(11, 11))
yrs = [2009,2010,2011,2012,2013]

ax1.plot(kpk_year,kpk_total)
ax1.set_ylabel('Area in million hectares', fontsize=12)
ax1.set_xlabel('Year', fontsize=12)
ax1.set_title('Total Area Irrigated in Khyber Pakhtunkhwa')

ax2.plot(punjab_year,punjab_total)
ax2.set_ylabel('Area in million hectares', fontsize=12)
ax2.set_xlabel('Year', fontsize=12)
ax2.set_title('Total Area Irrigated in Punjab')

ax3.plot(balochistan_year,balochistan_total)
ax3.set_ylabel('Area in million hectares', fontsize=12)
ax3.set_xlabel('Year', fontsize=12)
ax3.set_title('Total Area Irrigated in Balochistan')

ax4.plot(sindh_year,sindh_total)
ax4.set_ylabel('Area in million hectares', fontsize=12)
ax4.set_xlabel('Year', fontsize=12)
ax4.set_title('Total Area Irrigated in Sindh')

fig.subplots_adjust(wspace=0.5)
plt.show()
```
